```python

## inheritance -> other class can inherits from another class ,so if class B with inheritance from class A
# also gets inherits by class C , that class will inherits BOTH class A and B
# A -> B(A) -> C(B) = C(B,A) 
class Grandpa:
    lastName = 'ancestorson' #class variable;every class inherited,object instantiate from this class will have this lastname
    def __init__(self,Name):
        self.Name = 'gerald'
        
    def theWar(self):
        print('war never changes')
    
class pappy(Grandpa):
    def theDepression(self):
        print('and then the market crashed,along with every skulls against the floor')
        
class Sonny(pappy):
    def theRot(self):
        print('SIX SEVENNNNNNNN')
        
sunny = Sonny('sunny')
sunny.theRot()
sunny.theDepression()
sunny.theWar()
print(sunny.lastName)
```

```python
##abstract -> class that cannot be instantiate(made into object). meant to be a blueprint for subclass
from abc import ABC,abstractmethod #abstract based class

class enemyBlueprint(ABC):
    
    def __init__(self,health,damage,speed):
        self.health = health
        self.damage = damage
        self.speed = speed
        
    @abstractmethod
    def attack(self):
        print(f'hit for {self.damage} damages')
    
    @abstractmethod
    def pathfind(self):
        print('calculating path to target')
    
    #normal method test
    def death(self):
        print('defeated')
    
##cant make an object like this without implementing abstract methods  
# evilSoldier = enemyBlueprint()
## TypeError: Can't instantiate abstract class enemyBlueprint without an implementation for abstract methods 'attack', 'pathfind'

class zombieGrunt(enemyBlueprint):
    def attack(self):
        print(f'grahhhhhh {self.damage} damage')
    
    def pathfind(self):
        print('shambling to the target')
        
    #no neeed to implement death method,no abstract

# super() extend method of the parent class,just one line and boom; all functionality copied
class skeletonSoldier(enemyBlueprint):
    
    def __init__(self, health, damage, speed , defense):
        super().__init__(health, damage, speed)
        self.defense = defense
    
    def attack(self):
        return super().attack()
    
    def pathfind(self):
        return super().pathfind()

    def death(self):
        return super().death()

branzii = zombieGrunt(50 , 20, 10) 
branzii.attack()
branzii.pathfind()
varka = skeletonSoldier(70 , 30 , 20 , 10) 
varka.attack()
varka.pathfind()
varka.death()
```

```python
#polymorphism -> can be many thing all at once
from abc import ABC,abstractmethod

class shape():
    @abstractmethod
    def area():
        pass
    
class circle():
    
    def __init__(self,radius):
        self.radius = radius
    
    def area(self):
        return (f'{3.14 * self.radius} cm^2')

class pizza(circle):
    
    def __init__(self, radius,topping,size):
        super().__init__(radius)
        self.topping = topping
        self.size = size
        self.price = radius * 10
    
hawaiianPizza = pizza(20,'hawaiian','large')
print(hawaiianPizza.area())
print(hawaiianPizza.price)
##with this hawaiian pizza is now a pizza,a circle and a shape
```

```python
#duck type -> polymorphism by having minimum nescessary attribute or methods as other class

class animal():
    alive = True
    
class dog(animal):
    def speak (self):
        print('its the mutha f****** D-O-double g')

class wolf(animal):
    def speak (self):
        print('si esta meurto')
        
class roboRex(): #roborex is its own class but it has same func and attribute so it is animal(not really)
    
    alive = False
    
    def speak(self):
        print('*synth roar noise*') 

animals = [ dog() , wolf() , roboRex()]

for i in animals:
    i.speak()
    print(i.alive)
```

```python
##aggregation -> an object(a whole) contains a parts or refer to other object(a part)
class Library():
    def __init__(self,name):
        self.name = name
        self.bookList = []
    
    def add_book(self,book):
        self.bookList.append(book)
    
    def genreDump(self):
        return [f"book '{book.name}' genre is {book.genre}" for book in self.bookList]

class Book():
    def __init__(self,name,genre):
        self.name = name
        self.genre = genre
        
library = Library('placeholder library')
book1 = Book('the fold','mysteries')
book2 = Book('beginner guide to break and entering','comedies')

library.add_book(book1)
library.add_book(book2)
print([book.name for book in library.bookList]) #put book.name stuff in [] cuz its a list
for book in library.genreDump():
    print(book)
```

```python
# composition -> simmilar to aggregation but the object wholely owns its component 
                # component cant exist without that main object
                
class Engine():
    def __init__(self,horsepower):
        self.horsepower = horsepower

class Wheel():
    def __init__(self,size):
        self.size = size

class Car:
    def __init__(self , make , model , horsepower , size):
        self.make = make
        self.model = model
        self.engine = Engine(horsepower)
        self.wheels = [Wheel(size) for wheel in range(4)] # we can have only 4
    
    def display_car(self):
        return f'{self.make} {self.model}'

FordMustang = Car(make = 'ford',model = ' mustang' ,horsepower = 500 , size = 10)
FordMustang.display_car()
```
